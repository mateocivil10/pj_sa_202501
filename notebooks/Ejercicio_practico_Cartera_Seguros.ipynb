```python
import numpy as np
import matplotlib.pyplot as plt

# Definir parámetros
total_simulaciones = 10000  # Número de simulaciones
lambda_poisson = 1  # Tasa de ocurrencia de siniestros
media_gamma = 2000000  # Media de la severidad
std_gamma = 1000000  # Desviación estándar de la severidad

# Calcular los parámetros de la distribución Gamma
theta = std_gamma**2 / media_gamma  # Escala
k = media_gamma / theta  # Forma

# Simulación de Monte Carlo
costos_totales = []
for _ in range(total_simulaciones):
    num_siniestros = np.random.poisson(lambda_poisson)  # Número de siniestros
    costos = np.random.gamma(k, theta, num_siniestros)  # Costos de cada siniestro
    costo_total = np.sum(costos)  # Costo total de la cartera en la simulación
    costos_totales.append(costo_total)

# Convertir a array para cálculos
costos_totales = np.array(costos_totales)

# Mostrar estadísticas
print(f"Costo promedio esperado: {costos_totales.mean():,.2f}")
print(f"Percentil 95 del costo: {np.percentile(costos_totales, 95):,.2f}")

# Graficar la distribución de costos
plt.figure(figsize=(10, 5))
plt.hist(costos_totales, bins=50, alpha=0.75, edgecolor='black')
plt.xlabel("Costo Total de la Cartera")
plt.ylabel("Frecuencia")
plt.title("Distribución de Costos Simulados de la Cartera de Seguros")
plt.grid()
plt.show()
```
